**Checkpoint 2/7: Wordpiece Tokenizer**

How would the model we've been working with, i.e., `distilbert-base-uncased-finetuned-sst-2-english`, split this sentence? To examine this: 

- Create a tokenizer, `tokenizer`, for the model checkpoint `distilbert-base-uncased-finetuned-sst-2-english` using the `AutoTokenizer()` method. 
- Use the `.tokenize()` method on the sample sentence, to obtain `tokenized_array_wordpiece`, the set of tokens generated by the model and print it.


Don't forget to run the cell and save the notebook before selecting `Test Work`! Open the `Jupyter Help` toggle at the top of the notebook for more details.

```python
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"

## YOUR SOLUTION HERE ##
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
tokenized_array_wordpiece = tokenizer.tokenize(sample_sentence)

print(tokenized_array_wordpiece)
```

<details><summary style="display:list-item; font-size:16px; color:blue;"> Examining the tokens! </summary>

Examine the tokens generated by Wordpiece. Note how more commonly occurring words such as "is", "the" and even "soul" are preserved intact whereas the less common word "brevity" is broken into subwords. This makes sense if you think about the training corpus of a tokenizer and the frequency of words occurring. 
    
The biggest advantage of subword tokenization is that it can account for words that may not be in the training corpus, i.e., the so-called "rare words", in this manner.

**Checkpoint 6/7: Padding**

Create and print `model_inputs_padded_3` with the following arguments to the tokenizer: `padding=”max_length”` and `max_length=8`.

Don't forget to run the cell and save the notebook before selecting `Test Work`! Open the `Jupyter Help` toggle at the top of the notebook for more details.

```python
## Padding
# Will pad the sequences up to the maximum sequence length
model_inputs_padded_1 = tokenizer(sequences, padding="longest")

# Will pad the sequences up to the model max length
# (512 for BERT or DistilBERT)
model_inputs_padded_2 = tokenizer(sequences, padding="max_length")


## YOUR SOLUTION HERE ##
# Pad the sequences up to the specified max length
model_inputs_padded_3 = tokenizer(sequences, padding="max_length", max_length = 8)

print(model_inputs_padded_3)
```

**Checkpoint 7/7: Truncation**

```python
sequences = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

# Will truncate the sequences that are longer than the model max length
# (512 for BERT or DistilBERT)
model_inputs_truncated_1 = tokenizer(sequences, truncation=True)
print(model_inputs_truncated_1)

## YOUR SOLUTION HERE ##

model_inputs_truncated_2 = tokenizer(sequences, truncation=True, max_length = 8)
print(model_inputs_truncated_2)
```
